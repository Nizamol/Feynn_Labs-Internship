```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define parameters for revenue growth over time
months = np.arange(1,13)  #time period in months
initial_premium_subsribers = 500  #initial number of premium subscribers
growth_rate = 0.20  #growth in premium subscribers per month (assume there is a 20% growth)
P_subscription = 300 #price of subscription in INR
R_ads = 2000  #monthly advertisement revenue in INR
R_partnerships = 30000  #monthly partnership revenue in INR
```

```python
# Calculate the number of premium subscribers over time
premium_subscribers = initial_premium_subsribers * (1 + growth_rate)**months
premium_subscribers
```

```python
# Calculate total revenue for each month
total_revenue = (P_subscription * premium_subscribers) + R_ads + R_partnerships
total_revenue
```

```python
# Create the graph for total revenue over time
plt.figure(figsize=(10,6))
plt.plot(months, total_revenue, linestyle='-')
plt.title('Total Revenue Over Time')
plt.xlabel('Time Period (Months)')
plt.ylabel('Total Revenue (INR)')
plt.legend()
plt.show()
```
